```python
X = ev.drop(columns = ['Brand','Model','RapidCharge','PlugType','PriceEuro'])   #independent variables 
Y = ev['PriceEuro'] # dependent
```

```python
X_nor = pd.DataFrame(X,columns=['AccelSec','TopSpeed_KmH','Range_Km','Efficiency_WhKm','FastCharge_KmH','Seats'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.35,random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
lrmodel = LinearRegression()
```

```python
lrmodel
```

```python
lrmodel.fit(X_train,Y_train)
```

```python
k_result = lrmodel.predict(X_test)
```

```python
k_result
```

```python
Y_test
```

```python
X_test
```

```python
acc = lrmodel.score(X_test,Y_test)
```

```python
acc
```

```python
import pickle
pickle.dump(lrmodel,open('EVcar_model','wb'))
```
